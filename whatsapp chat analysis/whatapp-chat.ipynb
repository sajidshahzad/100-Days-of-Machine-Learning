```python
import pandas as pd
import numpy as np
import re
```

### opening text file

```python
f = open('_chat.txt', 'r', encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data) #expressions
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{1,2}:\d{1,2}\s([AP][M])]\s'
pattern1 = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{1,2}:\d{1,2}'
```

```python
message = re.split(pattern, data)[2::2]
message
```

```python
dates = re.findall(pattern1, data)
dates
```

```python
df = pd.DataFrame({'user_message': message, 'message_date': dates})
df['message_date'] = pd.to_datetime(df['message_date'], format ='%d/%m/%Y, %H:%M:%S')
df.rename(columns = {'message_date': 'date'}, inplace = True)
```

```python
df
```

```python
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notificatons')
        messages.append(entry[0])
df['user'] = users
df['message'] = messages
df['message'] = df['message'].str.strip('[')
df.drop(columns=['user_message'], inplace = True)
```
