```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
iris_setosa = df[df['species'] == 'setosa']
iris_versicolor = df[df['species'] == 'versicolor']
iris_virginica = df[df['species'] == 'virginica']
```

```python
df.head()
```

```python
sns.distplot(iris_setosa['petal_length'])
```

```python
sns.FacetGrid(df, hue='species', height= 6).map(sns.histplot,'petal_length', kde= True).add_legend()
```

```python
sns.FacetGrid(df, hue='species', height= 6).map(sns.histplot,'sepal_width', kde= True).add_legend()
```

```python
sns.FacetGrid(df, hue='species', height= 6).map(sns.histplot,'petal_width', kde= True).add_legend()
```

```python
cunts, bin_edges = np.histogram(iris_setosa['petal_length'], bins = 10)
```

```python
pdf = cunts/sum(cunts)
cdf = np.cumsum(pdf)
```

```python
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

```python
plt.style.use('fivethirtyeight')
```

```python
counts , bin_edges = np.histogram(iris_setosa["petal_length"], bins=10)
pdf = counts/sum(counts)
cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)


counts, bin_edges = np.histogram(iris_versicolor['petal_length'], bins =10)
pdf = counts/sum(counts)
cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)

counts, bin_edges = np.histogram(iris_virginica['petal_length'], bins =10)

pdf = counts/sum(counts)
cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
```
