# Standardization

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm
draw = norm.ppf(np.random.random(100000), loc =1000, scale = 200).astype(int)
sns.displot(draw, kde=True)
```

```python
z = []
for i in draw:
    z.append((1-1000)/200)
```

```python
a = np.array(z)
print(np.round((a.mean())))
```

```python
sns.displot(z)
```
