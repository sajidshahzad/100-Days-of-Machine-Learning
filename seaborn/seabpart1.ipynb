# 1. Scatter Plots

The scatter plot is a mainstay of statistical visualization. It depicts the distribution of the joint distribution of two variables using a cloud of points, where each point represents an observation in the dataset. This depiction allows the eye to infer a substantial amount of information about whether there is any meaningful relationship between them. 

Although, scatter plot can be plotted for both numerical as well as categorical data, it's mostly used to depict relationship between numerical data.

There are **2 ways** to draw a scatterplot in seaborn 

1. **relplot**[Figure Level Function]
2. **scatterplot**[Axes Level Function]

As we will see, these visualizations can be quite helpful beacuse they use simple and easily understood representations of data even on complex data by mapping up to three additional variables using the semantics of **hue, size, and style.**

```python
# importing libraires
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
data.head()
```

## Plot a Scatter Plot between total_bill and tip

```python
sns.relplot(x='total_bill', y='tip',data=data)
```

### The hue parameter

```python
sns.relplot(x='total_bill', y='tip', hue='smoker',data=data)
```

### The Style parameter

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', style='sex',data=data)
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', style='sex', size='size',data=data)
```
