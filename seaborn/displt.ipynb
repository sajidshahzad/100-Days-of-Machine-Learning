# Distplots

    are used to do univatiate analusis on a contious/numerical data.

**Distplot combines three things in one graph:**

  1. Histogram 
  2. Kernel Density Enstimation(PDF)
  3. Rugplot

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(), inplace= True)
```

```python
sns.displot(titanic['age'], bins= 10 , kde= True)
```

```python
sns.displot(titanic['age'], hist=True, rug=True)
```

```python
sns.displot(titanic[titanic['survived']==1]['age'],kde= True)
sns.displot(titanic[titanic['survived']==0]['age'],kde= True)
```
