```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Catagorical Scatter 


```python
# x --- Catagorical data
# y --- Numerical data

plt.style.use('fivethirtyeight')
```

```python
tip = sns.load_dataset('tips')
```

```python
sns.catplot(x='day', y='tip', kind='strip', data= tip) #jitter=1
```

## Swarm Plot

```python
sns.catplot(x='day', y='tip', kind='swarm', data= tip)
```

```python
sns.catplot(x='day', y='tip', kind='swarm',hue='sex', data= tip)
```

# Box Plot 

    is a standardize way of displaying distribution of data based on five number summary ('minimune' first quartile (Q1),median, third quatile (Q3) and the maximune). It can tell you about your outliers and what their values are. it can also tell you if your data is symmetrical, how tightly yur data is grouped, and if and how your data is skewed.

```python
sns.boxplot(tip['total_bill'])
```

```python
sns.catplot(x='day', y='total_bill', kind='box', data=tip)
```

```python
sns.catplot(x='day', y='total_bill', hue='sex', kind='box', data=tip)
```

# violine Plots
    violine plots are a method of plotting numeric data and can be considered a combination of the box plot with a kernel density plot.
    the unquestinable advantage of the violine plot over the box plot is that aside from showing the abovementioned  statistics it also shows the entire distribution of the data.
    violin plots are often used to compare the distribution of a given variable across some categories.

```python
sns.violinplot(tip['total_bill'])
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', data=tip)
```

```python
sns.catplot(x='day', y='total_bill', hue='sex', kind='violin', data=tip, split= True)
```

## Bar Plot and Countplot

```python
sns.catplot(x='smoker', y='total_bill', kind='bar', data= tip)
```

```python
sns.catplot(x='smoker', y='total_bill', hue='sex',kind='bar', data= tip)
```

```python
sns.catplot(x='sex', hue='smoker', kind='count', data= tip)
```

## Heatmaps

     is graphical representation of data where the individual vlaues contained in a matrix as represented as colors.

```python
flight = sns.load_dataset('flights')
flight.head()
```

```python
x = flight.pivot_table(index= 'year', columns='month', values='passengers', aggfunc='sum')
```

```python
sns.heatmap(x, linewidths=0.3, annot=True, fmt='d')
```

## Clustermap

```python
sns.clustermap(x)
```
