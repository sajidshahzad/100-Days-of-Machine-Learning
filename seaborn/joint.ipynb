# Joint Plot 
    creates amulti-panel figure that shows both the bivariate (or joint) relationship between two variables along with the univariate (or marginal) distribution of each on separate axes.


 _**we can use JointPlots in the following 4 manners:**_

     1. Scatter Plots
     2. Hexibin Plots 
     3. Kde Plots 
     4. Reg Plots

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tip = sns.load_dataset('tips')
plt.style.use('fivethirtyeight')
```

```python
sns.jointplot(x='total_bill', y='tip', data= tip )
```

```python
sns.jointplot(x='total_bill', y='tip', kind='hex', data= tip )
```

```python
sns.jointplot(x='total_bill', y='tip', kind= 'kde',data= tip )
```

```python
sns.jointplot(x='total_bill', y='tip', kind='reg', data= tip )
```
