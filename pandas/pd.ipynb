```python
delivery = pd.read_csv('deliveries.csv')
delivery.head()
```

```python
runs = delivery.groupby('batsman') #For top five lead scorers
```

```python
runs['batsman_runs'].sum().sort_values(ascending=False).head()
```

```python
mask = delivery['batsman_runs'] == 6  # top boundries by a batsman
new_del = delivery[mask]
fours = new_del.groupby('batsman')
fours['batsman_runs'].count().sort_values(ascending=False).head()
```

```python
def run_scored(batsman):
    vk = delivery[delivery['batsman'] == batsman]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

```python
run_scored("DA Warner")
```

# Merging two dataframes

```python
new_df = delivery.merge(df,left_on='match_id', right_on='id')
```

```python
new_df.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset = 'season' , keep='first').sort_values('season')[['batsman','season']]
```

## Pivot Table

```python
mask2 = delivery['batsman_runs'] == 6
six = delivery[mask2]
pt = six.pivot_table(index='over', columns='batting_team', values='batsman_runs', aggfunc='count')

```

```python
import seaborn as sns 
sns.heatmap(pt)
```
